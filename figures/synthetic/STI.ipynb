```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
import os
import pandas as pd
import numpy as np
import plotly.figure_factory as ff
```

```python
lengths = ["300bp", "3000bp", "30000bp"]
aln_ids = ['200580_114946_573911', '197113_332182_17210', '758_443154_73021', '198257_206396_13724']
names = ["High JSD, Low Entropy", "High JSD, High Entropy", "Low JSD, Low Entropy", "Low JSD, High Entropy"]
```

```python
figs = {}

for j, aln in enumerate(aln_ids):
    fig = make_subplots(
        rows=1, cols=3, 
        x_title="Theoretical Quantiles",
        y_title='Sample Quantiles',
        subplot_titles=("(a)", "(b)", "(c)"),
        shared_yaxes=True,
        )   
    for i, l in enumerate(lengths):
        data = pd.read_csv(f"~/repos/results/aim_3/synthetic/squartini_arndt/{aln}/{l}/STI.tsv", sep="\t")
        sti_1 = data["STI1"]
        sti_2 = data["STI2"]
        sti_3 = data["STI3"]

        hist_data = [sti_1, sti_2, sti_3]

        group_labels = ['STI 1', 'STI 2', 'STI 3']
        colors = ['#333F44', '#37AA9C', '#94F3E4']

        fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors)
        
        fig.update_layout(
            font_family="Helvetica",
            font_color="black",
            title_font_family="Helvetica",
            height=600, 
            width=600,
            )
            
        fig.show()
```
