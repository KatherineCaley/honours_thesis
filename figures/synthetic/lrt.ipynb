```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
import pandas as pd
import numpy as np
```

```python
lengths = ["300bp", "3000bp", "30000bp"]
aln_ids = ['200580_114946_573911', '197113_332182_17210', '758_443154_73021', '198257_206396_13724']
names = ["High JSD, Low Entropy", "High JSD, High Entropy", "Low JSD, Low Entropy", "Low JSD, High Entropy"]
```

```python
figs = {}

for j, aln in enumerate(aln_ids):
    name = names[j]
    fig = make_subplots(
        rows=1, cols=3, 
        x_title="Theoretical Quantiles",
        y_title='Sample Quantiles',
        subplot_titles=("(a)", "(b)", "(c)"),
        shared_yaxes=True,
    )

    for i, l in enumerate(lengths):
        data = pd.read_csv(f'/Users/katherine/repos/results/aim_2/synthetic/{aln}/{l}/pvals.tsv', sep="\t")
        p = data.loc[:,'p-value'].values
        p.sort()

        uniform_data = np.random.uniform(0,1,len(data))
        uniform_data.sort()

        # adding trace for qq, and then the diagonal line
        fig.add_trace(go.Scatter(x=uniform_data, y=p, line={"color": "mediumturquoise"}), row=1, col=i+1)
        fig.add_trace(go.Scatter(y=[0, 1], mode="lines", line={"color": "yellowgreen"}), row=1, col=i+1)
        
        # Update title and height
    fig.update_layout(height=400, width=800, showlegend=False)
    
    fig.layout.annotations[0].update(x=0.03, y=0.98)
    fig.layout.annotations[1].update(x=0.38, y=0.98)   
    fig.layout.annotations[2].update(x=0.72, y=0.98)       

    fig.update_layout(
    font_family="Times New Roman",
    font_color="black",
    title_font_family="Times New Roman",
    )
    figs[f"{aln}"]=fig
```

```python
for k, f in figs.items():

    # pio.write_image(f, f"plots/synthetic/lrt/{k}.pdf", format='pdf')
    f.show()
```

```python
# data for 1st one 
data1 = pd.read_csv(f'/Users/katherine/repos/results/aim_2/synthetic/197113_332182_17210/300bp/pvals.tsv', sep="\t")
pvals_column1 = data1.loc[:,'p-value']
pvals1 = pvals_column1.values
pvals1.sort()

# data for 2nd one 
data2 = pd.read_csv('/Users/katherine/repos/results/aim_2/synthetic/197113_332182_17210/30000bp/pvals.tsv', sep="\t")
pvals_column2 = data2.loc[:,'p-value']
pvals2 = pvals_column2.values
pvals2.sort()

# uniform data

uniform_data = np.random.uniform(0,1,1000)
uniform_data.sort()
```
